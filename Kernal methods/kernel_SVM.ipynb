## Kernal methods

```python
#create the gram matrix
import numpy as np

data_points = np.array([[1,5],[3,4],[4,2],[10,12]])

def kernel(x,z):
    return (1+ x[0]*z[0]+ x[1]*z[1])**2

def gram_matrix(data):

    gram = np.zeros((data.shape[0], data.shape[0]))
    for i in range(data.shape[0]):
        for j in range(data.shape[0]):
            gram[i][j] = kernel(data[i],data[j])
    return gram

g = gram_matrix(data_points)
print(g)

```

```python
# Find eigenvalues
eigenvalues,_ = np.linalg.eig(g)

print("Eigenvalues:")
print(eigenvalues)
```

```python
import matplotlib . pyplot as plt
from sklearn.datasets import make_circles

# Generate data with make_circles
np.random.seed(5)
X, y = make_circles(n_samples =500 , factor =0.3 , noise =0.1)
```

```python
X.shape
```

```python
plt.scatter(X[:,0],X[:,1], c=y, cmap = plt.cm.Paired)
```

```python
def map_X1(data):

    a = np.zeros((data.shape[0], 3))
    a[:,:2]= data
    a[:,2] = data[:,0]**2 + data[:,1]**2
    return a

X_new = map_X1(X)
print(X_new.shape)
```

```python
# Create a 3D scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot
ax.scatter(X_new[:, 0], X_new[:, 1], X_new[:, 2], c= y,cmap=plt.cm.Paired)

# Set labels for the axes
ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.set_zlabel('$x1^2 + x2^2$')
ax.view_init(15, 45)
# Show the plot
plt.show()
```
